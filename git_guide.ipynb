6 Main Step to push code to repo:
- Step 1: Create repo link in git/azure 
- Stept 2: git init (~ to git repo trong my ti working copy) 
- Step 3: git add . (~ add ton b file trong source vo staging area) => g git status  kim tra
- Step 4: git commit -m "write notes" (~ y code vo local repository) => g git log  kim tra
- Step5 : git remote add origin https://github.com/TrinhNKL/how_to_push_sourcecode_togit.git (~ ng b gia git trong local repository  commit vi repo trn git/azure trn server/mng --> nhm khai bo ng dn  push code ln git)
- Step 6: git push -u origin master (~ bc cui cng push code ln master)
