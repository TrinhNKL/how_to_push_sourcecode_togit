6 Main Step to push code to repo:
- Step 1: Create repo link in git/azure 
- Stept 2: git init (~ tạo git repo trong máy tại working copy) 
- Step 3: git add . (~ add toàn bộ file trong source vào staging area) => gõ git status để kiểm tra
- Step 4: git commit -m "write notes" (~ đẩy code vào local repository) => gõ git log để kiểm tra
- Step5 : git remote add origin https://github.com/TrinhNKL/how_to_push_sourcecode_togit.git (~ đồng bộ giữa git trong local repository đã commit với repo trên git/azure trên server/mạng --> nhằm khai báo đường dẫn để push code lên git)
- Step 6: git push -u origin master (~ bước cuối cùng push code lên master) 

Main step to create branch and push code:
- Step 1: git branch <branch name> --> ex: git branch develop (~ Tạo tên branch) => gõ git branch để kiểm tra các branch hiện có
- Step 2: git checkout <branch nam> --> ex git checkout develop (~ Chuyển nhánh sang develop)
- Step3 : git add. (~add toàn bộ file trong develop đã chỉnh sửa) -> gõ git status kiểm tra
- Step 4: git commit -m "write file to develop branch" (~commit lại code để chuẩn bị push)
- Step 5: git push origin develop (~ push code lên remote url)
